## 1. **Média (Esperança Matemática)**
A média de uma variável aleatória \(X\) é dada por:

- **Média de uma amostra**: 
  \[
  \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
  \]

- **Média de uma variável aleatória (discreta)**:
  \[
  \mu = E(X) = \sum_{x} x \cdot P(X = x)
  \]

- **Média de uma variável aleatória (contínua)**:
  \[
  \mu = E(X) = \int_{-\infty}^{\infty} x \cdot f_X(x) \, dx
  \]

## 2. **Variância**
A variância mede a dispersão dos valores em relação à média. É dada por:

- **Variância de uma amostra**:
  \[
  S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})^2
  \]

- **Variância de uma variável aleatória**:
  \[
  \sigma^2 = E[(X - \mu)^2] = \sum_{x} (x - \mu)^2 \cdot P(X = x)
  \]

  ou, equivalentemente,
  \[
  \sigma^2 = E(X^2) - [E(X)]^2
  \]

## 3. **Desvio Padrão**
O desvio padrão é a raiz quadrada da variância:

\[
\sigma = \sqrt{\sigma^2}
\]

## 4. **Separatrizes**
Separatrizes são medidas que dividem um conjunto de dados ordenados em partes iguais:

- **Quartis**: Dividem os dados em quatro partes iguais.
  - \(Q_1\) (1º quartil) — 25% dos dados abaixo deste valor.
  - \(Q_2\) (2º quartil ou mediana) — 50% dos dados abaixo deste valor.
  - \(Q_3\) (3º quartil) — 75% dos dados abaixo deste valor.

- **Decis**: Dividem os dados em dez partes iguais.
  - \(D_1\) (1º decil) — 10% dos dados abaixo deste valor.
  - \(D_9\) (9º decil) — 90% dos dados abaixo deste valor.

- **Percentis**: Dividem os dados em cem partes iguais.
  - \(P_1\) a \(P_{99}\) — \(P_k\) indica que \(k\%\) dos dados estão abaixo deste valor.

## 5. **Distribuição Normal**
A distribuição normal é uma distribuição contínua simétrica em torno da média, definida por:

- **Função densidade de probabilidade (PDF)**:
  \[
  f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
  \]

- **Propriedades**:
  - Média \( \mu \)
  - Desvio padrão \( \sigma \)
  - Curva simétrica em torno da média

## 6. **Distribuição Binomial**
Modela o número de sucessos em \(n\) ensaios independentes de Bernoulli, com probabilidade de sucesso \(p\):

- **Função de probabilidade**:
  \[
  P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
  \]
  onde \( \binom{n}{k} = \frac{n!}{k!(n-k)!} \)

## 7. **Distribuição Poisson**
A distribuição Poisson modela o número de eventos ocorrendo em um intervalo fixo de tempo ou espaço, com uma taxa média \( \lambda \):

- **Função de probabilidade**:
  \[
  P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
  \]

## 9. **Coeficiente de Variação**
O coeficiente de variação (CV) é uma medida de dispersão relativa, dada por:

\[
CV = \frac{\sigma}{\mu} \times 100\%
\]
